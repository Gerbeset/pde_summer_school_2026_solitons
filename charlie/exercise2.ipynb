## 2 Dispersive Wave Equations

We modify the wave equation by adding an additional term that depends on $u$:

$$
\partial_t^2 u - c^2\partial_x^2 u + g(u) = f(t,x),
$$

$$
u(0,x) = u^0(x),
$$

$$
\partial_t u(0,x) = v^0(x).
$$

**Example 1 (Klein–Gordon).** Choosing $g(u) = u$, $f(t,x) = 0$ gives us the Klein–Gordon equation. This is a linear dispersive PDE, where the wave speed of the solutions depends on its frequency. This can be seen by the following computation. Assume that $u(t,x) = A \exp(i(kx - \omega t))$ is a plane wave solution to the Klein–Gordon equation with wavenumber $k$ frequency $\omega$, and amplitude $A$. Inserting this into the PDE gives us the relation

$$
-\omega^2 + c^2k^2 + 1 = 0,
$$

which is called the dispersion relation for the Klein–Gordon equation. Dividing both sides by $k^2$ and rearranging gives us

$$
v^2 = c^2 + \lambda^2,
$$

where $v := \omega/k$ is the wave velocity and $\lambda := 1/k$ is the wavelength. This dispersion relation tells us that plane wave solutions of the Klein–Gordon equation with different wavelengths travel at different velocities, with shorter wavelengths traveling slower than longer wavelengths. In contrast, the dispersion relation for the wave equation with $g(u) = 0$, $f(t,x) = 0$ is

$$
v^2 = c^2,
$$

which means that all plane wave solutions to the wave equation travel at the same speed $c$.

For the general case, we decompose a wave solution $u$ of the Klein–Gordon equation into its fundamental frequencies

$$
u(t,x) = \sum_{k=-\infty}^{\infty} A_k e^{ik(x - v_k t)} + B_k e^{ik(x + v_k t)},
$$

with $v_k^2 = c^2 + \lambda_k^2$, $\lambda_k = 1/|k|$. As time evolves, the plane waves with wavelength $\lambda_k$ will travel faster than the plane waves with wavelength $\lambda_j$ for $j \leq k$, causing the initial profile of the solution $u(0,x)$ to spread out (i.e. disperse) over time.




### 2.1 Discretization

We use the scheme we developed for the wave equation with a small modification for the nonlinear term $g(u)$:

$$
\frac{U_i^{n+1} - 2U_i^n + U_i^{n-1}}{\tau^2} - c^2 \frac{U_{i+1}^n - 2U_i^n + U_{i-1}^n}{h^2} + g(U_i^n) = f_i^n
$$

We terminate all simulations when the wave fronts reach the boundary of the domain, so the choice of boundary condition does not matter much for our simulations. They only need to give a stable and well-posed linear system at the discrete level. The periodic boundary conditions from the previous section accomplish this. Another option is to use an absorbing boundary condition. At the lowest order approximation, this amounts to simply setting

$$
U_0^{n+1} = U_1^n,
$$

$$
U_M^{n+1} = U_{M-1}^n
$$

for $n \geq 1$.


### Finite Scheme Derivation
$$
\frac{U_i^{n+1} - 2U_i^n + U_i^{n-1}}{\tau^2} - c^2 \frac{U_{i+1}^n - 2U_i^n + U_{i-1}^n}{h^2} + g(U_i^n) = f_i^n \\
\frac{U_i^{n+1} - 2U_i^n + U_i^{n-1}}{\tau^2} = c^2 \frac{U_{i+1}^n - 2U_i^n + U_{i-1}^n}{h^2} - g(U_i^n) + f_i^n \\
U_i^{n+1} - 2U_i^n + U_i^{n-1} = \tau^2 c^2 \frac{U_{i+1}^n - 2U_i^n + U_{i-1}^n}{h^2} - \tau^2 g(U_i^n) + \tau^2 f_i^n \\

1. Extend the finite difference scheme for the wave equation to allow for $f(t,x)$ and $g(u)$, and the ability to choose either periodic or absorbing boundary conditions.



```python
def solve_wave_equation(
    L = 1,
    T = 40,
    c = 1,
    u_0 = lambda x: np.exp(-(x**2)),
    v_0 = lambda x,c: np.zeros_like(x),
    f = lambda x, t, u: np.zeros_like(x),
    g = lambda u: np.zeros_like(u),
    Boundary_condition = "absorbing"
    M = 200
    
    ):
    '''
    solve wave eq = f(t,x) on (a,b) x (0,T)
    '''

    #Initialize grids, vectors, and matrices
    h = (2*L) / (M + 1)
    N = int(c * T / h)
    tau = T / (N + 1)
    xs = np.array([-L + i * h for i in range(M + 2)])
    ts = np.array([n * tau for n in range(N + 2)])
    U = np.zeros((M + 2, N + 2))
    F = np.zeros((M+2, N+2))

    #Fill in F
    X, TT = np.meshgrid(xs, ts, indexing="ij")
    F = f(X, TT)

    # Initial condition at t = 0
    U[:, 0] = u_0(xs)
    U[M + 1, 0] = U[0, 0]

    #First time step
    U[1:M+1, 1] = (U[1:M+1, 0] + tau * v_0(xs[1:M+1],c)+ ((c**2 * tau**2) / (2 * h**2)) * (U[2:M+2, 0] - 2 * U[1:M+1, 0] + U[0:M, 0])+ (tau**2 / 2) * F[1:M+1, 0])
    U[0, 1] = (U[0, 0]+ tau * v_0(xs[0],c)+ ((c**2 * tau**2) / (2 * h**2)) * (U[1, 0] - 2 * U[0, 0] + U[M, 0])+ (tau**2 / 2) * F[0, 0])
    U[M + 1, 1] = U[0, 1]

    # Finite Difference Scheme:
    for n in range(1, N + 1):
        U[1:M+1, n+1] = (2 * U[1:M+1, n] - U[1:M+1, n-1] + ((c**2 * tau**2) / (h**2)) * (U[2:M+2, n] - 2 * U[1:M+1, n] + U[0:M, n])+ tau**2 * F[1:M+1, n])
        U[0, n+1] = (2 * U[0, n]- U[0, n-1]+ ((c**2 * tau**2) / (h**2)) * (U[1, n] - 2 * U[0, n] + U[M, n])+ tau**2 * F[0, n])
        U[M + 1, n+1] = U[0, n+1]


    return U, xs, ts



```
